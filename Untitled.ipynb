```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("winequality-red.csv", sep=";")
```

```python
print(df.head())
```

```python
df["quality"].value_counts().sort_index().plot(kind="bar", color="red")
plt.title("Wine Quality Distribution")
plt.xlabel("Quality Score")
plt.ylabel("Number of Wines")
plt.show()
```

```python
df.boxplot(column="alcohol", by="quality")
plt.title("Alcohol vs Wine Quality")
plt.xlabel("Quality")
plt.ylabel("Alcohol %")
plt.show()
```

```python
corr = df.corr()
plt.imshow(corr, cmap="coolwarm")
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Heatmap")
plt.show()
```

```python

# Trend 1: High quality wines mein alcohol % zyada hota hai
high_quality = df[df['quality'] >= 7]
low_quality = df[df['quality'] <= 4]
print(f"High Quality avg alcohol: {high_quality['alcohol'].mean():.2f}")
print(f"Low Quality avg alcohol:  {low_quality['alcohol'].mean():.2f}")

# Trend 2: Volatile acidity = quality ka dushman
print(f"High Quality avg volatile acidity: {high_quality['volatile acidity'].mean():.3f}")
print(f"Low Quality avg volatile acidity:  {low_quality['volatile acidity'].mean():.3f}")

# Trend 3: Sulphates = quality ka dost
print(f"High Quality avg sulphates: {high_quality['sulphates'].mean():.3f}")
print(f"Low Quality avg sulphates:  {low_quality['sulphates'].mean():.3f}")
```

```python
print("\n========== RISKS ==========")
print("R1: Volatile acidity > 0.8 → wine tastes like vinegar ")
print("R2: Alcohol < 9.5% → weak body, low quality")
print("R3: High volatile acidity + low alcohol = guaranteed bad quality")

print("\n========== OPPORTUNITIES ==========")
print("O1: Alcohol 10-12% range → move from average to premium ")
print("O2: Sulphates increase → better preservation, higher score")
print("O3: Control volatile acidity < 0.4 → quality jumps up")
```
